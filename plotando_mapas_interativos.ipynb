 ### Adicionando marcados
  Têm vários tipos de marcadores, começando com um Leafletmarcador de local de estilo simples com um pop-up e uma dica de ferramenta HTML.

```python
mapa = folium.Map(
    location=[-5.811193, -35.206101],
    zoom_start=15,
)

folium.Marker(
    location=[-5.811193, -35.206101],
    popup='<i> Midway Mall</i>',
    tooltip='Click Aqui!',
    icon=folium.Icon(color='green')

).add_to(mapa)


folium.Marker(
    location=[-5.811867, -35.203918],
    popup='<i> IFRN </i>',
    tooltip='Click Aqui!',
    icon=folium.Icon(color='red')

).add_to(mapa)

folium.Circle(
    radius=100,
    location=[-5.811867, -35.203918],
    popup='The Waterfront',
    color='crimson',
    fill=False,
).add_to(mapa)



folium.CircleMarker(
    location=[-5.811867, -35.203918],
    radius=50,
    popup='Laurelhurst Park',
    color='#3186cc',
    fill=True,
    fill_color='#3186cc'
).add_to(mapa)

mapa
```

Exibindo as coordenadas de Latitude e Longitude no evento de click

```python
mapa.add_child(folium.LatLngPopup())
```

## Visualizando dados de Vendas por no RN

```python
import pandas as pd
from folium import plugins
```

```python
df = pd.read_csv("/home/weslley/Documentos/Untitled Folder/base/olist_geolocation_dataset.csv")
```

```python
df.head()
```

```python
df=df[df['geolocation_state']=='RN']
#df.geolocation_city.unique()
df.geolocation_city.value_counts().head(5)


#df.info()
```

```python
df.geolocation_city.value_counts().head()
```

```python
df.describe()
```

```python
coordenadas=[]
for lat,lng in zip(df.geolocation_lat.values[:],df.geolocation_lng.values[:]):
    coordenadas.append([lat,lng])
```

```python
mapa = folium.Map(
    location=[-5.811867, -35.203918],
    zoom_start=7,
    )
```

```python
mapa.add_child(plugins.HeatMap(coordenadas))        
mapa
```
